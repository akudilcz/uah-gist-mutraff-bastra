```python
PATH="data"
SCENARIO="GRID64_6F"

SCENARIO="Madrid_Retiro"
SCENARIO="Madrid_Tablas"
SCENARIO="Madrid_Salamanca"
```

```python
file_nodes="{}/{}_nodes.csv".format(PATH,SCENARIO)
file_edges="{}/{}_edges.csv".format(PATH,SCENARIO)
file_net="{}/{}_network.csv".format(PATH,SCENARIO)
file_flows="{}/{}_od.csv".format(PATH,SCENARIO)

outfile = "{}.net".format(SCENARIO)
```

```python
def isNaN(num):
    return num != num
```

```python
def translate_2_maslab( scenario, outfile, file_nodes, file_edges, file_net, file_flows ):

    low_memory=False
    
    sep=","
    fh = open(outfile, 'w')
    fh.write("#################################\n")
    fh.write("# Autogenerated file for scenario: {}\n".format(scenario))
    fh.write("#################################\n")
    fh.write("#function name (args) formula\n")
    fh.write("function OW (f) t+0.02*f \n")
    fh.write("#################################\n")
    fh.write("#node name \n")
    df = pd.read_csv(file_nodes, encoding='utf-8', sep=sep)
    for index, row in df.iterrows():
        v=row[0];
        if(type(v)==numpy.float64):
            fh.write("node N{} \n".format(int(row[0])))
        else:
            fh.write("node N{} \n".format(row[0]))
        
        
    dfe = pd.read_csv(file_edges, encoding='utf-8', sep=sep )
    dfe.columns.values[0] = "node_name"
    dfn = pd.read_csv(file_net, encoding='utf-8', sep=sep)
    dfn.columns.values[0] = "node_name"
    nr=len(dfe.columns)
    hnodes={}
    inodes={}
    for x in range(1,nr):
        nname='N{}'.format(dfe.columns[x])
        hnodes[nname]=x
        inodes[x-1]=nname
    fun="OW"

    fh.write("#################################\n")
    fh.write("#edge name origin destination function constants \n")
    for x, rowData in dfe.iterrows():
        nc = len(rowData)
        for y in range(1,nc):
            if(not isNaN(rowData[y])):
                ename = 'E{}'.format(rowData[y]).replace('#','_')
                #print("edge {} {} {}".format(cell,x,y))
                fh.write("edge {} {} {} {} {} \n".format(ename,inodes[x],inodes[y-1],fun,dfn.iloc[x,y]))

    fh.write("#################################\n")
    fh.write("#od name origin destination flow\n")
    dff = pd.read_csv(file_flows, encoding='utf-8', sep=sep )
    for x, rowData in dff.iterrows():
        nc = len(rowData)
        for y in range(1,nc):
            cell = rowData[y]
            if(not isNaN(cell) and cell>0):
                #print("edge {} {} {}".format(cell,x,y))
                fh.write("od {}|{} {} {} {} \n".format(inodes[x],inodes[y-1],inodes[x],inodes[y-1],cell))


    fh.close()
```

```python
translate_2_maslab( SCENARIO, outfile, file_nodes, file_edges, file_net, file_flows)
```

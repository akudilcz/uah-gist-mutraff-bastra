### load_tazs
Read a TAZ file and prepare a dictionary for use

```python
def load_tazs( taz_file):
    tazs={}
    xtree = et.parse(taz_file)
    xroot = xtree.getroot()
    for node in xroot: 
        taz_id = node.attrib.get("id")
        tazs[taz_id]={}
        tazs[taz_id]['id']=taz_id

        if 'edges' in node.attrib:
            taz_edges = node.attrib.get("edges")
            tazs[taz_id]['edges']=taz_edges.split(' ')
        else:
            tazs[taz_id]['edges']=[]
        # print( "TAZ[{}]={}".format(tazs[taz_id]['id'],tazs[taz_id]['edges']))
        # s_age = node.find("age").text
    return tazs
```

### complete_tazs_with_ids
Given a TAZS structure, create a edge-id list similar to the edge_name list to filter data

```python
def complete_tazs_with_ids( tazs, exp_edges ):
    edge_name_to_id={}
    for index, row in exp_edges.iterrows():
        edge_name_to_id[row['id']]={}
        edge_name_to_id[row['id']]['id']=row['name']
    #print( edge_name_to_id )

    for key in tazs:
        edge_ids=[]
        if 'edges' in tazs[key]:
            # print(tazs[key]['edges'])
            for name in tazs[key]['edges']:
                try:
                    edge_ids.append( edge_name_to_id[name]['id'] )
                except:
                    print("WARNING - EDGE {} WAS FOUND IN TAZS BUT NOT IN EDGES".format(name))
                    edge_ids.append( '???' )
            # print( edge_ids )
            tazs[key]['edge_ids']=edge_ids
        # print(tazs[key])
    return tazs
```

```python
# Example: 
# draw_graph( data, "Total halted vehicles", "traf_halted_veh_num", "Halted vehicles", "sum" )
def create_graph( title, measure_name, measure_label, num_points, y_max ):

    # Prepare the plot
    fig, axes = plt.subplots(figsize=(15,7))

    plt.xlabel('Simulation time', fontsize=BIG_FONT_SIZE)
    plt.ylabel(measure_label,fontsize=BIG_FONT_SIZE)

    # BLACK STYLE
    axes.spines['bottom'].set_color('0.5')
    axes.spines['top'].set_color('0.5')
    axes.spines['right'].set_color('0.5')
    axes.spines['left'].set_color('0.5')
    axes.patch.set_facecolor('0.1')
    plt.grid(b=True, which='major', color='0.2', linestyle='-')
    plt.grid(b=True, which='minor', color='0.2', linestyle='-')
    axes.tick_params(axis='x', colors='0.7', which='both')
    axes.tick_params(axis='y', colors='0.7', which='both')
    axes.yaxis.label.set_color('0.9')
    axes.xaxis.label.set_color('0.9')
    axes.margins(0.5)
    plt.title( title, color='0.7' )
    fig.patch.set_facecolor('0.15')

        
    # WHILE STYLE
    axes.spines['bottom'].set_color('0')
    axes.spines['top'].set_color('0')
    axes.spines['right'].set_color('0')
    axes.spines['left'].set_color('0')
    axes.patch.set_facecolor('1')
    plt.grid(b=True, which='major', color='0.2', linestyle='-')
    plt.grid(b=True, which='minor', color='0.2', linestyle='-')
    axes.tick_params(axis='x', colors='0.7', which='both')
    axes.tick_params(axis='y', colors='0.7', which='both')
    axes.yaxis.label.set_color('0.2')
    axes.xaxis.label.set_color('0.2')
    axes.margins(0.5)

    axes.set_xlim( xmin=0, xmax=num_points )

    plt.title( title, color='0.2' )
    fig.patch.set_facecolor('1')
    
    plt.rc('font', size=MID_FONT_SIZE)          # controls default text sizes
    plt.rc('axes', titlesize=MID_FONT_SIZE)     # fontsize of the axes title
    plt.rc('axes', labelsize=BIG_FONT_SIZE)    # fontsize of the x and y labels
    plt.rc('xtick', labelsize=SMALL_FONT_SIZE)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=SMALL_FONT_SIZE)    # fontsize of the tick labels
    plt.rc('legend', fontsize=BIG_FONT_SIZE)    # legend fontsize
    plt.rc('figure', titlesize=BIG_FONT_SIZE)  # fontsize of the figure title
    
    return fig, axes
```

### plot_bar
Draw horizontal bar diagram

```python
def plot_bar(title,xlabel,ylabel,labels,values):
    # this is for plotting purpose
    index = np.arange(len(labels))
    plt.bar(index, values)
    plt.xlabel(xlabel, fontsize=10)
    plt.ylabel(ylabel, fontsize=10)
    plt.yticks(index, labels, fontsize=14, rotation=30)
    plt.title(title)
    plt.show()
```

### plot_barh
Draw vertical bar diagram

```python
def label_bars(ax, bars, text_format, **kwargs):
    """
    Attaches a label on every bar of a regular or horizontal bar chart
    """
    ys = [bar.get_y() for bar in bars]
    y_is_constant = all(y == ys[0] for y in ys)  # -> regular bar chart, since all all bars start on the same y level (0)

    if y_is_constant:
        _label_bar(ax, bars, text_format, **kwargs)
    else:
        _label_barh(ax, bars, text_format, **kwargs)


def _label_bar(ax, bars, text_format, **kwargs):
    """
    Attach a text label to each bar displaying its y value
    """
    max_y_value = ax.get_ylim()[1]
    inside_distance = max_y_value * 0.05
    outside_distance = max_y_value * 0.01

    for bar in bars:
        text = text_format.format(bar.get_height())
        text_x = bar.get_x() + bar.get_width() / 2

        is_inside = bar.get_height() >= max_y_value * 0.15
        if is_inside:
            color = "white"
            text_y = bar.get_height() - inside_distance
        else:
            color = "black"
            text_y = bar.get_height() + outside_distance

        ax.text(text_x, text_y, text, ha='center', va='bottom', color=color, **kwargs)


def _label_barh(ax, bars, text_format, **kwargs):
    """
    Attach a text label to each bar displaying its y value
    Note: label always outside. otherwise it's too hard to control as numbers can be very long
    """
    max_x_value = ax.get_xlim()[1]
    # distance = max_x_value * 0.0025

    refval = bars[0].get_width()
    # print("BARS: ref={}, max={}".format(refval, max_x_value))

    for bar in bars:
        val = bar.get_width()
        percent = (val-refval)*100/refval
        text = text_format.format(bar.get_width())+" / {:.2f}%".format(percent)

        text_x = bar.get_width() - 10
        text_x = 0
        text_y = bar.get_y() + bar.get_height() / 2

        ax.text(text_x, text_y, text, va='center', color='white', weight='bold', fontsize=12, **kwargs)
```

```python
def plot_barh(title,xlabel,ylabel,labels,values):
    fig, ax = plt.subplots()
    
    # this is for plotting purpose
    index = np.arange(len(labels))
    bars = ax.barh(index, values)
    plt.ylabel(xlabel, fontsize=10)
    plt.xlabel(ylabel, fontsize=10)
    # plt.set_yticks(index, labels, fontsize=14, rotation=30)
    plt.yticks(index, labels, fontsize=14)
    plt.title(title)
    
    value_format = "{:.2f}"  # displaying values as percentage with one fractional digit
    label_bars(ax, bars, value_format)
    plt.show()
```

```python
def plot_barh_edge_measures( exps, target_TAZ, label_X, col_name, units, scale_x ):
    label_Y = "Scenario"
    labels=[]
    values=[]

    if target_TAZ == "":
        filter_TAZ = False
    else:
        filter_TAZ = True
        edges=tazs[target_TAZ]['edge_ids']
        # print("Selected EDGES={}".format(edges))

    for exp in exps:
        # df1=exps[2]['data']
        # df1[df1['edge_id'].isin(taz_edges)].head()
        if filter_TAZ:
            # print(exp['data'].columns)
            df = exp['data'][ exp['data']['edge_id'].isin(edges)]
            # display(HTML("Selected {} rows ({} cols)".format(df.shape[0],df.shape[1])))
            accum = df[col_name].sum()/scale_x
        else:
            accum = exp['data'][col_name].sum()/scale_x
            # display(HTML("Selected ALL data: {} rows ({} cols)".format(exp['data'].shape[0],exp['data'].shape[1])))
        labels.append(exp['label'])
        values.append(accum)
        # print( "{} : {:02f} {} in {}".format(measure,accum,units,exp['label']))
    # print(labels)
    # print(values)
    plot_barh(label_X,label_Y,units,labels,values)
```
